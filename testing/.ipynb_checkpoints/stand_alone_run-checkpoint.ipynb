# Tutorial: run an exported model

This notebook is a short tutorial for running an exported `.pt2` model.

`y_next_phys = model(y_phys, dt_seconds, g_phys)`

The example stays on CPU and uses a small batch so the call is vectorized.


## Step 1: load the exported model

This first step loads the model and reads the embedded metadata so the species order, global order, and usable `dt_seconds` range are visible.


```python
from pathlib import Path
import json

import torch

EXPORT_PATH = Path("export_cpu_dynB_1step_phys.pt2")
DEVICE = torch.device("cpu")
DTYPE = torch.float32

extra_files = {"metadata.json": ""}
exported_program = torch.export.load(str(EXPORT_PATH), extra_files=extra_files)
metadata = json.loads(extra_files["metadata.json"])
model = exported_program.module().to(device=DEVICE, dtype=DTYPE)

species_names = list(metadata["species_variables"])
global_names = list(metadata["global_variables"])
dt_min_seconds = metadata["dt_min_seconds"]
dt_max_seconds = metadata["dt_max_seconds"]

print(f"export: {EXPORT_PATH.resolve()}")
print(f"species order: {species_names}")
print(f"global order:  {global_names}")
print(f"dt range from training/preprocessing: {dt_min_seconds:.6e} to {dt_max_seconds:.6e} seconds")
```

## Step 2: build a small batch of inputs

This step creates three example rows.

- `y_phys` is the species input
- `dt_seconds` is the timestep in seconds
- `g_phys` is the global input

For this tutorial, keep `dt_seconds` inside the printed range above. That range is the one baked into the export from training/preprocessing.


```python
batch_size = 3
n_species = len(species_names)
n_globals = len(global_names)

# y_phys shape: (batch_size, n_species)
# Start from a simple positive composition, then make each row a little different.
y_phys = torch.full((batch_size, n_species), 1.0 / n_species, device=DEVICE, dtype=DTYPE)
y_phys[torch.arange(batch_size), torch.arange(batch_size) % n_species] += 0.10
y_phys = y_phys / y_phys.sum(dim=1, keepdim=True)

# dt_seconds shape: (batch_size,)
# Use a few values across the dt range that was used in training/preprocessing.
dt_seconds = torch.linspace(
    dt_min_seconds,
    dt_max_seconds,
    steps=batch_size,
    device=DEVICE,
    dtype=DTYPE,
)

# g_phys shape: (batch_size, n_globals)
# In this tutorial, P and T are filled in by hand.
g_phys = torch.zeros((batch_size, n_globals), device=DEVICE, dtype=DTYPE)
g_phys[:, global_names.index("P")] = 1.0e6
g_phys[:, global_names.index("T")] = torch.tensor([900.0, 1000.0, 1100.0], device=DEVICE, dtype=DTYPE)

print("y_phys:")
print(y_phys)
print()
print("dt_seconds:")
print(dt_seconds)
print()
print("g_phys:")
print(g_phys)
```

## Step 3: run the model

This is the full inference call for the batch.


```python
with torch.inference_mode():
    y_next_phys = model(y_phys, dt_seconds, g_phys)

print(f"output shape: {tuple(y_next_phys.shape)}")
print()

for batch_index in range(batch_size):
    print(f"batch item {batch_index}")
    for name, value in zip(species_names, y_next_phys[batch_index].tolist()):
        print(f"  {name:>20s} : {value:.6e}")
    print()
```
